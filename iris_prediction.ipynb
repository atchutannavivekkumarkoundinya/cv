```python
#libraries
import pandas as pd 
import seaborn as sns #to colour data points
import matplotlib.pyplot as plt #to draw graph
import numpy as np 
from sklearn.datasets import load_iris  
```

```python
iris = pd.read_csv("iris.csv") #imported data 
```

```python
print(iris.shape) #to know how many data points and features
```

```python
iris["variety"].value_counts() 
# to know how how many data points are alloted to each type 
#each batch has same number of data points 50,50,50 so this type of data set is called as balanced data set
#we have total of 150 data points and 3 differnt flowers so it is divided into 50,50,50
```

```python
print(iris.columns) #to know how many columns names in the data set
```

```python
iris.plot(kind = 'scatter', x = 'sepal.length',y = 'sepal.width');
#it is a scatter plot to represent sepal length in x axis and sepal width in y axis
#we are reprenting with a scatter point 
plt.show();#it show the plot we are showing
```

```python
#this is used to colour and differentiate the data points using variety of flowers 
# it is used to coloue the data points using variety coloumn
#it is used to give a white grid in the output graph 
# it is used to polt sepal length and sepal width in x and y aaxis respectively
#it is used to give  colour diffinition for out in the graph
sns.set_style("whitegrid");
sns.FacetGrid(iris, hue = 'variety', size = 5) \
   .map(sns.scatterplot,"sepal.length","sepal.width") \
   .add_legend();
plt.show();
```

```python
#this graph is used to plot 3-d scatter points
#as we have different data point which cannot be represent in 2 d
# so we use pairplot which compares 2 different data points and plots the graph
sns.set_style("whitegrid");
sns.pairplot(iris , hue= 'variety',size = 3,diag_kind="kde");
plt.show()
#it has only 4c2 = 6 unqiue graph
```

```python
sns.FacetGrid(iris, hue = 'variety', size = 5) \
   .map(sns.distplot, "petal.length") \
   .add_legend();
plt.show();

```

```python
sns.FacetGrid(iris, hue = 'variety', size = 5) \
   .map(sns.distplot, "petal.width") \
   .add_legend();
plt.show();
```

```python
sns.FacetGrid(iris, hue = 'variety', size = 5) \
   .map(sns.distplot, "sepal.length") \
   .add_legend();
plt.show();
```

```python
sns.FacetGrid(iris, hue = 'variety', size = 5) \
   .map(sns.distplot, "sepal.width") \
   .add_legend();
plt.show();
```

BY THIS WE CAN SAY PETAL LENGTH > PETAL WIDTH > SEPAL LENGTH > SEPAL WIDTH
